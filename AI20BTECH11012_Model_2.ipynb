# **Deleting Columns which have more than 11000 Null Values in the train_df and test_df**

```python
####
# Storing the Columns which have more than 11000 Null Values in train_df in the below list

high_null_columns = []
for column in train_df.columns:
    if(train_df[column].isnull().sum() > 11000):
        high_null_columns.append(column)
```

```python
####
# Printing the highly null columns

print(high_null_columns)
```

# **The columns in high_null_columns have many null values(>11000) so we can't delete samples having null values (leads to decrase in training samples) and we can not fill the null values with mode of that feature (leads to skewed features).**

# **So, we drop the features which are in high_null_columns from train_df and test_df.**

```python
####
# Dropping high_null_columns from train_df amd test_df 

train_df.drop(columns=high_null_columns, axis=1, inplace=True)
test_df.drop(columns=high_null_columns, axis=1, inplace=True)
```

# **The feature 'Location' in train_df and test_df contains two values separated by space or ','.**

# **We have to split that 'Location' feature into two new features 'Location 1' and 'Location 2' in train_df and test_df.**

```python
####
# Splitting 'Location' in train_df into two new columns 'Location 1' and 'Location 2'

train_df[['Location 1','Location 2']] = train_df['Location'].str.split(' |,',expand=True)
```

```python
####
# Splitting 'Location' in test_df into two new columns 'Location 1' and 'Location 2'

test_df[['Location 1','Location 2']] = test_df['Location'].str.split(' |,',expand=True)
```

```python
####
# Dropping 'Location' Column from train_df as we already split the data in it

train_df.drop(columns=["Location"], axis=1, inplace=True)
```

```python
####
# Dropping 'Location' Column from test_df as we already split the data in it

test_df.drop(columns=["Location"], axis=1, inplace=True)
```

# **Now, we fill the Null Values in columns having Null Values with the mode of that Column.**

```python
####
# Storing the Columns having Null Values in train_df in the below list

null_columns_train = []

for column in train_df.columns:
    if(train_df[column].isnull().sum() != 0):
        null_columns_train.append(column)
```

```python
####
# Prining null_columns in train_df

null_columns_train
```

```python
####
# Storing the Columns having Null Values in test_df in the below list

null_columns_test = []

for column in test_df.columns:
    if(test_df[column].isnull().sum() != 0):
        null_columns_test.append(column)
```

```python
####
# Prining null_columns in train_df

null_columns_test
```

# **So, null_columns_train = null_columns_test .**

# **Now, we fill the Null Values in train_df and test_df with mode of that column having Null Values.**

```python
####
# Copying train_df, test_df into another data frames

train_data = train_df.copy()
test_data = test_df.copy()
```

```python
####
# Filling Null Values in train_data and test_data

for column in null_columns_train:
    train_data[column].fillna(train_data[column].mode()[0], inplace=True)
    test_data[column].fillna(test_data[column].mode()[0], inplace=True)
```

# **The columns 'Location 1' and 'Location 2' are of type 'object'.**

# **We have to type cast them into 'float' type.**

```python
####
# Casting 'Location 1' in train_data into float type

train_data["Location 1"] = train_data["Location 1"].astype(float)
train_data["Location 2"] = train_data["Location 2"].astype(float)
```

```python
####
# Casting 'Location 1' and 'Location 2' in test_data into float type

test_data["Location 1"] = test_data["Location 1"].astype(float)
test_data["Location 2"] = test_data["Location 2"].astype(float)
```

# **Now, we have to encode the Categorical Features into Numeric Data.**

# **Beforing encoding the Categorical Features, we convert the Categorical Features into 'category' type to decrease Time Complexity for encoding.**

```python
####
# We create a list to store the Categorical Features in the list

categorical_columns = train_data.select_dtypes(include='object').columns.to_list()

categorical_columns
```

```python
####
# Typecasting Categorical Columns into 'category' type

for column in categorical_columns:
    train_data[column] = train_data[column].astype('category')
    test_data[column] = test_data[column].astype('category')
```

```python
####
# Printing dtypes of each column in train_data

print(train_data.dtypes)
```

```python
####
# Printing dtypes of each column in train_data

print(test_data.dtypes)
```

```python
####
# Endoing the Categorical Columns into Numeric Data using cat.codes

for column in categorical_columns:
    train_data[column] = train_data[column].cat.codes
    test_data[column] = test_data[column].cat.codes
```

```python
####
# Printing Encoded train_data

train_data
```

```python
####
# Printing Encoded test_data

test_data
```

# **Now, we have to split train_data into train_x (contains only features) and train_y (contains only labels).**

```python
####
# Copying train_data into a new Data Frame

train_x = train_data.copy()
```

```python
####
# Extracting 'Fault' Column into train_y list

train_y = train_data['Fault'].to_list()

# Dropping 'Fault' Column from train_x Data Frame

train_x.drop(columns=["Fault"],axis=1,inplace = True)
```

```python
####
# Printing train_x (only features) data frame

train_x
```

```python
####
# Printing train_y (only labels) list

train_y
```

```python
####
# Copying test_data into a new Data Frame

test_x = test_data.copy()
```

```python
####
# Extracting 'Id' Column into test_id list

test_id = test_data["Id"].to_list()

# Dropping 'Id' Column from test_x Data Frame

test_x.drop(columns=["Id"],axis=1,inplace = True)
```

```python
####
# Printing test_id

test_id
```

```python
####
# Printing test_x (only labels) data frame

test_x
```

```python
#### Converting train_x, train_y, test_x into numpy arrays

train_x_arr = train_x.to_numpy()

train_y_arr = np.array(train_y)

test_x_arr = test_x.to_numpy()
```

# **Implementing the Model**

# **XG Boosting**

```python
####
# Importing XG Boost Classifier

from xgboost import XGBClassifier

# Applying Tuned Parameters

clf = XGBClassifier(learning_rate=0.05,max_depth=10,n_estimators=170)

# Training the Classifier

clf.fit(train_x_arr,train_y_arr)
```

```python
####
# Predicting test_x using the Classifier

predictions = clf.predict(test_x_arr)

print(predictions)
```

```python
####
# Attaching test_id to predictions and adding them in submission data frame

submission = pd.DataFrame()
submission['Id'] = test_id
submission['Fault'] = list(predictions)
```

```python
####
# Prnting the Submission

submission
```

```python
####
# Saving submission into a csv file

submission.to_csv('final_submission_2.csv',index=False)
```
